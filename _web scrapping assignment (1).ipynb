```python
import requests
```

```python
url="https://www.flipkart.com/search?q=led+tv&sid=ckf%2Cczl&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=led+tv%7CTelevisions&requestId=6c936b16-df62-4b93-bd37-cd0c0dfd0ccb&as-searchtext=led"
```

```python
res=requests.get(url)
```

```python
res
```

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

```python
tv_names=soup.select('._4rR01T')
```

```python
tv_names
```

```python
LED_tv=[]
```

```python
for i in tv_names:
    LED_tv.append(i.text)
    
```

```python
LED_tv
```

```python
ratings=soup.select('._3LWZlK')
```

```python
ratings
```

```python
ratings_list=[]
```

```python
for k in ratings:
    ratings_list.append(k.text)
```

```python
ratings_list
```

```python
reviews=soup.select('._34hpFu span span , ._13vcmD+ span')
```

```python
reviews
```

```python
reviews_list=[]
```

```python
for t in reviews:
    reviews_list.append(t.text)
```

```python
reviews_list
```

```python
price=soup.select('._1_WHN1')
```

```python
price
```

```python
price_list=[]
for item in price:
    price_list.append(item.text)
    
```

```python
price_list
```
